```python
hash = {}
count = 0
input_file = open('f.txt')
for line in input_file:
        num = int(line.rstrip('\n'))
        hash[num] = 1
input_file.close()

def target_sum (t):
        for element in hash:
                t_el = t-element
                if t_el in hash and t_el != element:
                        return 1
        return 0

for i in range(-10000,10000):
    if i % 100 == 0:
        print(i)
    if target_sum(i):
        count += 1
print(count)
```
